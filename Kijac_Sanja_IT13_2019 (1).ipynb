**O SEGMENTACIJI**

Segmentacija fotografije je proces razdvajanja slike na različite segmente ili regione radi identifikacije i analize specifičnih objekata ili regija unutar slike[1]. Cilj segmentacije je prepoznavanje granica između različitih objekata ili regiona kako bi se omogućilo razumevanje i tumačenje sadržaja slike na nivou pojedinačnih delova[1]. Segmentacija fotografije može se izvršiti na više nivoa detaljnosti, od grubog segmentiranja na velike regione do fino-granularne segmentacije na piksel nivou.

**VRSTE SEGMENTACIJE**

Osnovne vrste segmentacije su[2]:

1. **Segmentacija po boji:** Ova vrsta segmentacije koristi boje kao kriterijum za razdvajanje objekata na slici. Može se koristiti thresholding ili klasterovanje boja kako bi se odvojili objekti na osnovu njihovih boja ili raspona boja.

2. **Segmentacija po ivicama:** Ova vrsta segmentacije identifikuje ivice objekata na slici. Može se izvršiti korišćenjem algoritama detekcije ivica kao što su Canny detekcija ivica ili Sobel operator.

3. **Segmentacija po regijama:** Ova vrsta segmentacije razdvaja sliku na osnovu sličnosti piksela u određenom regionu. Algoritmi poput region rasta ili algoritma sa gradijentom (eng. region growing or region merging) se često koriste za grupisanje piksela u slične regione.

4. **Segmentacija po teksturi:** Ova vrsta segmentacije razdvaja objekte na slici na osnovu njihovih teksturalnih karakteristika. Koriste se metode analize teksture, kao što je Gaborov filter ili teksturni deskriptori, kako bi se identifikovali regioni sa sličnim teksturalnim osobinama.

5. **Segmentacija po pokretu:** Ova vrsta segmentacije se koristi za razdvajanje objekata na slici na osnovu njihovog kretanja. Koristi se u oblastima kao što su praćenje objekata, video analiza i robotsko viđenje.

6. **Semantička segmentacija:** Ova vrsta segmentacije dodeljuje semantičke oznake svakom pikselu na slici kako bi se identifikovali i razumeli različiti objekti ili regioni. Ova vrsta segmentacije se često koristi u oblasti kompjuterskog vida i autonomnih vozila. Kako izgledaju dve osnovne vrste segmentacije prikazano je na narednoj slici:

**TEHNIKE SEGMENTACIJE**

Postoje različite podele algoritama i tehnika za segmentaciju u zavisnosti od autora ali najveći broj njih se svodi na sledeće[3]:

1.   **Segmentacija pragom (eng. *Thresholding*)** - grupa metoda zasnovanih na poređenju osvetljenosti piksela sa jednim ili više pragova, pa tako imamo segmentaciju sa jednim i segmentaciju sa više pragova. Ukoliko imamo slike u boji tada se vrši poređenje boje između piksela[3].
2.   **Segmentacija klasterizacijom (eng. *Clustering*)** - primenjivana u segmentaciji multispektralnih satelitskih ili avionskih snimaka, gdje je osnovni cilj bio izdvajanje regiona na površini zemlje koji imaju iste karakteristike, kao što su npr., isti tip poljoprivrednih kultura, isti sastav zemljišta, ista nadmorska visina, itd. Nedostatak ove metode je što ima veliku računsku složenost[3].
3.   **Segmentacija regionom (eng. *Region*)** - predstavlja skup metoda koje pokušavaju da izdvoje područja slike koja su homogena sa gledišta određenih karakteristika[3].
4.   **Segmentacija granicom regiona (eng. *Edge*)** - zahteva da se primeni i neki postupak za spajanje ivica, kako bi se formirale neprekidne granice regiona iz razloga što metode koje vrše izdvajanje ivica generišu isprekidane granice objekata a ne zatvorene krive[3]. Glavna prednost ovog metoda je njena izuzetna jednostavnost i ona daje dobre rezultate u slučaju segmentacije jednostavnih scena. Međutim, kada imamo slike na kojima postoji veći broj objekata koji se preklapaju, mapa ivica sadrži tačke grananja, pa u tom slučaju ova metoda ne daje dobre rezultate.

U okviru projekta korišćena je U-Net konvoluciona neuronska mreža koja obezbeđuje brzu i preciznu semantičku segmentaciju slika[3]. Sastoji se od dva glavna dela: "enkoder" i "dekoder". U-Net arhitektura je postala popularna u oblasti segmentacije slika zbog svoje sposobnosti da nauči i rekonstruiše detaljne informacije iz slika, čak i kada su objekti mali ili slabo kontrastni. Naredna klasa olakšava manipulaciju skupom podataka za segmentaciju tako što omogućava pristup pojedinačnim slikama i maskama. Transformacije se primenjuju  tokom pristupa podacima, što je korisno za pripremu podataka za obuku neuronske mreže.

```python
class Segmentation_Dataset(Dataset):
    def __init__(self, image_path, mask_path):
        self.image_path = image_path
        self.mask_path = mask_path

        images = glob.glob(image_path + '/*.png')

        samples = []

        for image in images:
            samples.append(image.split('/')[-1].split('.')[0])

        self.samples = samples

        self.transforms = T.Compose([T.Grayscale(), T.ToTensor()])

    def __getitem__(self, idx):
        image = Image.open(os.path.join(self.image_path, self.samples[idx] + '.png'))
        mask = Image.open(os.path.join(self.mask_path, self.samples[idx] + '.png'))

        return self.transforms(image), self.transforms(mask)

    def __len__(self):
        return len(self.samples)
```

Klasa ConvolutionBlock predstavlja jedan konvolucijski blok u arhitekturi neuronske mreže. Ako je self.activation postavljen na True, tada se na rezultat primenjuje ReLU aktivacija, a krajnji rezultat je tenzor x s primenjenom ReLU aktivacijom.

Ako je self.activation postavljen na False, tada se ne primenjuje ReLU aktivacija i krajnji rezultat je samo tenzor x bez promene.

U oba slučaja, nakon primene konvolucije i normalizacije, krajnji rezultat je tenzor x.

```python
class ConvolutionBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, stride=1, activation=True):
        super(ConvolutionBlock, self).__init__()

        self.activation = activation

        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, stride=stride, padding=kernel_size // 2)

        self.norm = nn.BatchNorm2d(out_channels)

    def forward(self, x):
        x = self.conv(x)
        x = self.norm(x)

        if self.activation:
            return F.relu(x)
        else:
            return x
```

Klasa ResidualBlock predstavlja jedan rezidualni blok u arhitekturi neuronske mreže. Krajnji je da se vrati rezultirajući tenzor x kao izlaz iz forward metode ResidualBlock-a.

```python
class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3):
        super(ResidualBlock, self).__init__()

        self.norm = nn.BatchNorm2d(in_channels)

        self.conv1 = ConvolutionBlock(in_channels, out_channels, kernel_size)
        self.conv2 = ConvolutionBlock(out_channels, out_channels, kernel_size, activation=False)

    def forward(self, x):
        residual = x

        x = F.relu(x)

        x = self.norm(x)
        x = self.conv1(x)
        x = self.conv2(x)

        return x
```

Klasa DeconvolutionBlock predstavlja jedan dekonvolucijski blok u arhitekturi neuronske mreže. Krajnji rezultat ove konkatenacije je tenzor x koji ima povećane dimenzije iz tenzora xd i dodatnu dimenziju tenzora x2.


```python
class DeconvolutionBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=2, stride=2):
        super(DeconvolutionBlock, self).__init__()

        self.deconv = nn.ConvTranspose2d(in_channels, out_channels, kernel_size=kernel_size, stride=stride)

    def forward(self, x1, x2):
        xd = self.deconv(x1)

        x = torch.cat([xd,x2], dim=1)

        return x
```

Konačno, definisanje U-Net konvolucijske neuronske mreže koja će biti model i nad kojom će se vršiti treniranje. Prilikom definisanja modela kombinuju se konvolucijski i rezidualni blokovi kako bi se izgradila ResNet mreža.

```python
class UNet(nn.Module):
    def __init__(self, filters=16, dropout=0.5):
        super(UNet, self).__init__()

        self.conv1 = nn.Sequential(
            nn.Conv2d(1, filters, 3, padding=1),
            ResidualBlock(filters, filters),
            ResidualBlock(filters, filters),
            nn.ReLU()
        )

        self.conv2 = nn.Sequential(
            nn.MaxPool2d(2, 2),
            nn.Dropout(dropout / 2),
            nn.Conv2d(filters, filters * 2, 3, padding=1),
            ResidualBlock(filters * 2, filters * 2),
            ResidualBlock(filters * 2, filters * 2),
            nn.ReLU()
        )

        self.conv3 = nn.Sequential(
            nn.MaxPool2d(2, 2),
            nn.Dropout(dropout),
            nn.Conv2d(filters * 2, filters * 4, 3, padding=1),
            ResidualBlock(filters * 4, filters * 4),
            ResidualBlock(filters * 4, filters * 4),
            nn.ReLU()
        )

        self.conv4 = nn.Sequential(
            nn.MaxPool2d(2, 2),
            nn.Dropout(dropout),
            nn.Conv2d(filters * 4, filters * 8, 3, padding=1),
            ResidualBlock(filters * 8, filters * 8),
            ResidualBlock(filters * 8, filters * 8),
            nn.ReLU()
        )

        self.middle = nn.Sequential(
            nn.MaxPool2d(2, 2),
            nn.Dropout(dropout),
            nn.Conv2d(filters * 8, filters * 16, 3, padding=3 // 2),
            ResidualBlock(filters * 16, filters * 16),
            ResidualBlock(filters * 16, filters * 16),
            nn.ReLU()
        )

        self.deconv4 = DeconvolutionBlock(filters * 16, filters * 8, 2)

        self.upconv4 = nn.Sequential(
            nn.Dropout(dropout),
            nn.Conv2d(filters * 16, filters * 8, 3, padding=1),
            ResidualBlock(filters * 8, filters * 8),
            ResidualBlock(filters * 8, filters * 8),
            nn.ReLU()
        )

        self.deconv3 = DeconvolutionBlock(filters * 8, filters * 4, 3)

        self.upconv3 = nn.Sequential(
            nn.Dropout(dropout),
            nn.Conv2d(filters * 8, filters * 4, 3, padding=1),
            ResidualBlock(filters * 4, filters * 4),
            ResidualBlock(filters * 4, filters * 4),
            nn.ReLU()
        )

        self.deconv2 = DeconvolutionBlock(filters * 4, filters * 2, 2)

        self.upconv2 = nn.Sequential(
            nn.Dropout(dropout),
            nn.Conv2d(filters * 4, filters * 2, 3, padding=1),
            ResidualBlock(filters * 2, filters * 2),
            ResidualBlock(filters * 2, filters * 2),
            nn.ReLU()
        )

        self.deconv1 = DeconvolutionBlock(filters * 2, filters, 3)

        self.upconv1 = nn.Sequential(
            nn.Dropout(dropout),
            nn.Conv2d(filters * 2, filters, 3, padding=1),
            ResidualBlock(filters, filters),
            ResidualBlock(filters, filters),
            nn.ReLU(),
            nn.Dropout(dropout / 2),
            nn.Conv2d(filters, 1, 3, padding=1)
        )

    def forward(self, x):
        conv1 = self.conv1(x)
        # 101 -> 50
        conv2 = self.conv2(conv1)
        # 50 -> 25
        conv3 = self.conv3(conv2)
        # 25 -> 12
        conv4 = self.conv4(conv3)
        # 12 - 6
        x = self.middle(conv4)
        # 6 -> 12
        x = self.deconv4(x, conv4)
        x = self.upconv4(x)
        # 12 -> 25
        x = self.deconv3(x, conv3)
        x = self.upconv3(x)
        # 25 -> 50
        x = self.deconv2(x, conv2)
        x = self.upconv2(x)
        # 50 -> 101
        x = self.deconv1(x, conv1)
        x = self.upconv1(x)

        return x
```

IoU (Intersection over Union) je metrika koja se često koristi za evaluaciju performansi modela u problemu segmentacije objekata. IoU meri koliko se dobro predikcija modela podudara s referentnim ili stvarnim oznakama objekata. IoU ima vrednosti između 0 i 1, gde veća vrednost ukazuje na bolje podudaranje predikcije sa referentnom oznakom. Vrednost 1 znači savršeno preklapanje između predikcije i stvarne oznake, dok vrednost 0 ukazuje na potpuno nepodudaranje.

```python
def iou_score(outputs, labels):
    A = labels.squeeze().bool()

    pred = torch.where(outputs < 0., torch.zeros_like(outputs), torch.ones_like(outputs))

    B = pred.squeeze().bool()

    intersection = (A & B).float().sum((1,2))

    union = (A | B).float().sum((1,2))

    iou = (intersection + 1e-6) / (union + 1e-6)

    return iou
```

Ova funkcija predstavlja jednu iteraciju treninga nad jednom mini-batchom podataka. Ova funkcija se koristi tokom treninga kako bi se izvršila jedna iteracija (prosleđivanje ulaza kroz mrežu, izračunavanje gubitka, propagacija unazad, ažuriranje parametara).

```python
def train_one_batch(model, x, y):
    x, y = x.to(device), y.to(device)

    outputs = model(x)

    loss = loss_fn(outputs, y)
    iou = iou_score(outputs, y).mean()

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    return loss.item(), iou.item()
```

Ova funkcija se može koristiti za prikazivanje vremena proteklog od nekog trenutka kako bi se pratio napredak treninga. U ovom primeru se koristi za prikazivanje koliko vremena je proteklo od početka treninga.

```python
def time_since(since):
    now = time.time()

    s = now - since
    m = math.floor(s / 60)
    s -= m * 60

    return '%dm %ds' % (m,s)
```

Treniranje modela objašnjeno je u narednim tezama.

Prvo se postavljaju hiperparametri kao što su NUM_EPOCHS (broj epoha treninga) i BATCH_SIZE (veličina batch-a). Zatim se instancira objekat modela UNet i prebacuje se na odgovarajući uređaj na GPU.

Nakon toga, postavljaju se optimizer (torch.optim.Adam) i scheduler (torch.optim.lr_scheduler.ReduceLROnPlateau) za prilagođavanje stope učenja tokom treninga.

Definiše se funkcija gubitka (loss_fn) koja se koristi za izračunavanje gubitka modela.

Kreira se DataLoader (train_dataloader) koji će iterirati kroz trening dataset (train_dataset) koristeći određenu veličinu batch-a i nasumično mešanje podataka.

Zatim se prolazi kroz svaku epohu treninga i u svakoj epohi se prolazi kroz sve batch-eve podataka u DataLoader-u.

Za svaki batch, poziva se funkcija train_one_batch koja obavlja trening jednog batch-a podataka koristeći model, ulazne podatke (x) i oznake (y). Funkcija vraća gubitak i IoU rezultat za taj batch.

Nakon svakog batch-a, prikazuju se informacije o trenutnom gubitku i IoU-u za taj batch.

Na kraju svake epohe, prikazuju se prosečni gubitak, prosečni IoU i proteklo vreme za tu epohu.

Scheduler se koristi za prilagođavanje stope učenja na osnovu prosečnog IoU-a.

Takođe, tokom treninga se beleže gubitak (train_loss) i IoU rezultati (train_iou) za svaku epohu.

```python
NUM_EPOCHS = 150
BATCH_SIZE = 64

model = UNet().to(device)

model.train()

optimizer = torch.optim.Adam(model.parameters())
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max')

loss_fn = nn.BCEWithLogitsLoss()

train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)

steps = train_dataset.__len__() // BATCH_SIZE

start = time.time()

train_loss = []

train_iou = []

for epoch in range(1, NUM_EPOCHS+1):
    print('Epoch {}/{}'.format(epoch, NUM_EPOCHS))

    running_iou = []
    running_loss = []

    for step, (x,y) in enumerate(train_dataloader):
        loss, iou = train_one_batch(model, x, y)

        running_iou.append(iou)
        running_loss.append(loss)

        print('\r{:6.1f} %\tloss {:8.4f}\tIoU {:8.4f}'.format(100 * (step + 1) / steps, loss, iou), end='')

    print('\r{:6.1f} %\tloss {:8.4f}\tIoU {:8.4f}\t{}'.format(100 * (step + 1) / steps, np.mean(running_loss), np.mean(running_iou), time_since(start)))

    scheduler.step(np.mean(running_iou))

    train_loss.append(loss)
    train_iou.append(iou)
```

Nakon treniranja modela radi se grafički prikaz rasta/pada funkcije gubitka (engl. loss function) i IoU tokom izvršenih 150 iteracija.
Može se primetiti da IoU dostiže 0.8 što je solidan rezultat iako je poželjno da bude bliže 1. Da bi se poboljšao rezultat pretpostavka je podesiti nove parametre ili upotrebiti drugi model koji je bolji od ovog odnosno koji je ostvario bolje rezultate.

```python
plt.plot(train_loss, label='loss')
plt.plot(train_iou, label='IoU')

plt.xlabel('epoch')
plt.ylabel('metric')

plt.legend()

plt.show()
```

Nakon treniranja i evaluacije ostaje još testiranje. U ovom kodu se definiše funkcija rle_encode(im) koja vrši enkodiranje maske u RLE format. Ključni parametar je ulazna maska im, koja se transformiše u jednodimenzionalni niz piksela. Zatim se identifikuju serije piksela sa istom vrednošću i njihove dužine se izračunavaju. Konačni rezultat je RLE kodiranje dužina serija piksela koje se vraća kao string.

Ova funkcija omogućava kompresiju maski i efikasno čuvanje informacija o prisutnosti objekata. Takođe, kod uključuje definiciju objekta transforms koji se koristi za primenu transformacija na slikama, uključujući konverziju u grayscale i pretvaranje u tensor.

```python
def rle_encode(im):
    pixels = im.flatten(order='F')
    pixels = np.concatenate([[0], pixels, [0]])

    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1

    runs[1::2] -= runs[::2]

    return ' '.join(str(x) for x in runs)

transforms = T.Compose([T.Grayscale(), T.ToTensor()])
```

Ovaj kod prolazi kroz svaki red u CSV datoteci SAMPLE_SUBMISSION_PATH. Za svaki red izdvaja identifikator (id) i učitava odgovarajuću sliku iz testnog skupa koristeći putanju IMAGE_TEST_PATH i dodajući identifikator u ime fajla. Zatim primenjuje definisane transformacije na učitanu sliku. Slika se šalje modelu kako bi se dobio izlaz. Nakon toga, primenjuje se prag na izlazu kako bi se dobila binarna maska. Koristi se funkcija rle_encode da se kodira binarna maska u RLE format, i dodaje se rezultat u listu rle_mask.

Na kraju, se dodaje nova kolona 'rle_mask' u okviru podataka sub_df, koja sadrži kodirane RLE maske.

```python
sub_df = pd.read_csv(SAMPLE_SUBMISSION_PATH)

n = sub_df.shape[0]

rle_mask = []

for idx in range(n):
    sample = sub_df['id'][idx]

    image = Image.open(os.path.join(IMAGE_TEST_PATH + '/{}.png'.format(sample)))
    image = transforms(image)

    out = model(image.unsqueeze(0)).squeeze()

    pred = torch.where(out < 0., torch.zeros_like(out), torch.ones_like(out))

    rle_mask.append(rle_encode(pred.numpy()))

sub_df['rle_mask'] = rle_mask
```
